```python
import numpy as np
import pandas as pd
from collections import Counter
import string
import warnings
warnings.filterwarnings('ignore')


#This part is useful for getting the filename and location the first time but after that you don't need it
"""
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
"""
        
file = "data/internet_archive_scifi_v3.txt"

# should use context manager to open files but this doesn't work with Jupyter notebooks (which Kaggle is based on)
text = open(file,'r').read()
```

```python
toy_text = text[0:1000000]
toy_text
```

```python
# split text into words and count most common

lst = toy_text.split(" ")
print('Number of characters = ', len(lst), '\nNumber of words = ', len(lst), '\n')

sc = Counter(lst)
print(sc.most_common(20))
```

```python
# Lets turn everything to lowercase and try again

lower_lst = toy_text.lower().split(" ")
lower_sc = Counter(lower_lst)
print(lower_sc.most_common(20))
```

```python
# Some of those results are just punctuation so lets remove them.

text_snip = toy_text[0:1000]

print('Original has punctuation: \n', text_snip, '\n')

text_snip2 = text_snip.translate(str.maketrans('', '', string.punctuation))

print('This one does not:\n', text_snip2)
```

```python
# Lets put this all together and find out our sci-fi themes!

fixed_text = text.lower().translate(str.maketrans('', '', string.punctuation)).split(" ")
fixed_sc = Counter(fixed_text)

print(fixed_sc.most_common(20))
```

```python
# Zipf's Law

# It has been observed that if you count the frequency of words in a text 
# then multiply that by the rank order of the words they tend to be equal

import matplotlib.pyplot as plt

df = pd.DataFrame.from_dict(fixed_sc, orient='index')
df.sort_values(0, ascending=False, inplace=True)

df = df.rename(columns={None :'event', 0:'count'})
df['rank'] = df['count'].rank(method='max', ascending=False)
df['Zipfs'] = df['count'] * df['rank']

df.iloc[0:30].plot(y='Zipfs')
```

```python
# import matcher class
from spacy.matcher import Matcher

# create matcher object and pair it to the current vocab, which must be shared with the documents the matcher will operate on
matcher = Matcher(nlp.vocab)
```

```python
text6 = "Brazil plan travel ban on flights from Egypt. Brazil will ban flights from Egypt"

doc6 = nlp(text6)
```

```python
pattern = [{'ENT_TYPE': "GPE", 'OP': '+'}, {'IS_ALPHA': True, "OP": "*"}, {'LEMMA': 'ban', 'POS': 'VERB'},
           {'IS_ALPHA': True, "OP": "*"}, {'ENT_TYPE': "GPE", 'OP': '+'}]

# ENT_TYPE: "GPE" > searches for a country name
# OP: + > requires the pattern to match 1 or more times

# IS_ALPHA: True > token text consists of alphanumeric characters
# OP: + > requires the pattern to match 0 or more times

# LEMMA: 'ban' > search for the lemma of the word 'ban'
# POS: "VERB" > requires the parts of speech of the word ban to be a verb
```

```python
pattern
```

```python
# add the pattern to the matcher object with the name BanMatcher (you can call it whatever you want), and set callbacks to None
matcher.add('BanMatcher', [pattern])

# apply matcher on our Doc
matches = matcher(doc6)
print(matches)
```

```python
for match_id, start, end in matches:
    span = doc6[start:end]
    print(start, end, span.text) # print start and end position of the match, and the span text
```
